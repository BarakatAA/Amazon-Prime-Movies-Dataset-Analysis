### Data Cleaning Steps

```python
# removing columns that won't be needed in my analysis such as description and date_added
df.drop(['description', 'date_added'], axis = 1, inplace = True)
```

```python
#checking to see execution of previous command
df.head()
```

```python
#applying split function to get the genre a movie/tv show falls into from the listed-in column
df['genre'] = df.listed_in.str.split(pat =',', expand = True)[0]
```

```python
#splitting the cast colum to get the main_cast
df['main_cast'] = df.cast.str.split(pat = ',', expand = True)[0]
```

```python
df.head()
```

```python
# dropping the old listed_in and cast columns
df.drop(['listed_in', 'cast'], axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

```python
#splitting the country column to get the main_country because some movies have more than one country listed
df['main_country'] = df.country.str.split(pat = ',', expand = True)[0]
```

```python
df.drop(['country'], axis = 1, inplace = True)
```

### Exploratory Data Analysis

Question 1

How many Movies versus Tv Shows are on Amazon Prime?

```python
sns.countplot(x='type', data=df)
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Count of Movies and TV Shows')
plt.show()
```

This shows that there are more movies listed (over 7000) on Amazon Prime compared to around 2000 Tv shows

#### Question 2

Which genre is the most popular in Amazon Prime?


```python
def plot_cat(var, l=12, b=5):
    plt.figure(figsize = (l, b))
    sns.countplot(df[var], order = df[var].value_counts().index)
    plt.xticks(rotation = 90)
```

```python
plot_cat('genre')
```

This shows that the most popular genre is Drama followed by Action and Comedy while the least popular genre is Historical

#### Question 3

How many movies/Tv shows are available in each release year?

```python
plot_cat('release_year')
```

The most movies/tvshows (over 1400) were released in 2021 while the least movies were released in 1927

#### Question 6

Which directors had the most releases?

```python
genre_counts = df['director'].value_counts().head(10)

sns.countplot(x='director', data=df, order=genre_counts.index)
plt.xticks(rotation=90)
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Directors with most releases')
plt.show()
```

The plot shows that  Director Mark Knight had the most releases with a whooping 100+ releases to his credit! This gets me curious and I'd like to see what genres his movie/Tv show releases fall in

#### Question 7

What is the genre spread for our top director Mark Knight?

```python
# To isolate Mark Knight, the director with the most releases
top_director = df['director'].value_counts().index[0]

# Filtering the DataFrame for the movies directed by MK
movies_by_MK = df[df['director'] == top_director]

# Creating a count plot of genres for the movies by MK
sns.countplot(x='genre', data=movies_by_MK)
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title(f'Genre Distribution of Movies by {top_director}')
plt.show()
```

This shows that Mark Knight's movie/ show releases are spread across Fitness, Faith and Special interests to documentaries and music videos with the highest being Special interests. This makes empirical sense as a director of Action movies or sci fi and adventure movies can hardly have that much movie/tv show releases like Mark Knight because those categories are usually high in production budget with filming for a single movie spanning over 2 years. It is also likely Mark Knight's releases have a short duration and I'm curious to find this out.

#### Question 8

What is the duration spread for Mark Knight's releases? Does he have more short-lenght films, music videos and documentaries?

```python
top_director = df['director'].value_counts().index[0]

movies_by_MK = df[df['director'] == top_director]

sns.histplot(data=movies_by_MK, x='duration', bins=10)
plt.xticks(rotation = 90)
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.title(f'Duration Distribution of Movies by {top_director}')
plt.show()
```

From the above histogram plot, it is clear that Mark Kinght has a spread of releases with duration ranging from as low as 9 minutes to as high as 601 minutes with his highest number of releases in the less than 60 minute ballpark. This further lends credence to the reason he was able to amass such a high number of releases.
